**Task 1: Exploring and Visualizing a Simple Dataset**

***Objective:***

Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

```python
import pandas as pd          # For loading and exploring data
import seaborn as sns        # For loading Iris and making plots
import matplotlib.pyplot as plt   # For charts
```

***Dataset:***

Iris Dataset (CSV format,loaded via seaborn)

```python
df = sns.load_dataset("iris")
```

.

.

**○  Print the shape, column names, and the first few rows using .head().**

```python
print("Shape of the dataset (rows, columns):")
print(df.shape)
```

```python
print("\nColumn names:")
print(df.columns)
```

```python
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

.

.

**○  Use .info() and .describe() for summary statistics.**

```python
print("\nBasic info about each column:")
print(df.info())
```

```python
print("\nSummary statistics (min, max, mean, etc.):")
print(df.describe())
```

.

.

**Visualize the dataset:**

○ Create a scatter plot to show relationships between features.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df, 
    x="sepal_length",   # value on X-axis
    y="sepal_width",    # value on Y-axis
    hue="species"       # color by flower type
)
plt.title("Sepal Length vs Sepal Width (Scatter Plot)")
plt.show()
```

.

.

**○ Use histograms to show value distributions.**

```python
df.hist(figsize=(10,8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

.

.

**○ Use box plots to identify outliers.**

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Box Plots of Iris Dataset Features")
plt.show()
```

The results show that Iris is a small, clean, balanced dataset with 
clear separation between classes especially in petal features, making 
it ideal for beginner machine learning.
